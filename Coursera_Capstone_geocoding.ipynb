```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import time
```

```python
canada = pd.read_csv('./canada_postarea.csv')
```

```python
geocodes = pd.read_csv('./Geospatial_Coordinates.csv')
```

```python
geocodes.head()
```

```python
geocodes.rename(columns={'Postal Code': 'Postcode'}, inplace=True)
```

```python
canada = pd.merge(left=canada, right=geocodes, how='inner', on='Postcode')
```
